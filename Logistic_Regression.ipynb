# Logistic Regression
It is a type of linear classification to predict the categorical data (dependent variable) from a no. of non independent variable
We use Sigmoid Function for this

## Importing Libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

# Importing the DataSet

````python
dataset = pd.read_csv("Social_Network_Ads.csv")
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
````

## Diving data into testing and Training set

````python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
````

````python
print(np.concatenate((X_train,Y_train.reshape(len(Y_train),1)),1))
````

## Feature Scaling the model

````python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
````

## Logistic Regression of Data

````python
from sklearn.linear_model import LogisticRegression
classifer = LogisticRegression()
classifer.fit(X_train,Y_train)
````

## Predicting a new Resutlt


````python
print(classifer.predict(sc.transform([[44,39000]])))
````

## Predicing Test Set Results

````python
Y_pred = classifer.predict(X_test)
print(np.concatenate((Y_test.reshape(len(Y_test),1),Y_pred.reshape(len(Y_pred),1)),1))
````

## Confusion Matrix
The Confusion matrix will show us the exact number of correct predictions and Number of Incorrect predictions. It is a 2D matrix

```
 [ correct 0  incorrect 1]
 [  65            3      ]
 [  8             24     ]
 [ incorrect 0  correct 1]

```

  

````python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(Y_test,Y_pred)
s = accuracy_score(Y_test,Y_pred)
print(f"The confusion matrix is :\n {cm} ")
print(f"The accuracy score is : {s}")
````

##  Visualizing the Training Set Result

````python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), Y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifer.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
````

## Visulaizing the Test set Result

````python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), Y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.25))
plt.contourf(X1, X2, classifer.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
````
