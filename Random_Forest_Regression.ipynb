# Random Forest Regression

This model works on Ensemble Learning where we apply the same Decision Tree Intution Model a number of times say 10 and find the average from it to predict the result.
This improves the accuracy as the number of outcomes increases the probabily of correct result same as in the case of deciding the no of marbles in a jar by 100000 people and averaging it or finding the median gives the correct results.

# Importing the Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing the dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
Y = dataset.iloc[:,-1].values
```

# Training the model

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=20, random_state=0)
regressor.fit(X,Y)
```

# Visualizing the Result of Model

```python
X_grid = np.arange(min(X),max(X),0.01)
#print(X_grid)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X,Y,color='green')
plt.plot(X_grid,regressor.predict(X_grid),color='red')
plt.title("Level Vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

# Predicting the Result for 6.5 Level

```python
regressor.predict([[6.5]])
```
